### Household heterogeneity and the transmission of foreign shocks

#### **1. The paper and its model**

**Paper:** [Household heterogeneity and the transmission of foreign shocks](https://www.sciencedirect.com/science/article/pii/S0022199620300222)

**Authors:** Sergio de Ferra, Kurt Mitman, Federica Romei

**Journal:** Journal of International Economics (2020)

##### **The agent’s objective**

The authors develop a heterogeneous-agent New Keynesian small open economy model (HANKSOME) to study how household heterogeneity affects the transmission of foreign shocks, particularly during a current account reversal. 
The model highlights how households' portfolio composition and the extent of foreign currency borrowing shape the contraction in consumption following sudden stops in capital inflows.

Households maximize the expected present discounted value of utility:
$$
U = E_0 \sum_{t=0}^{\infty} \beta^t \left[ u(c_t) - g(l_t) \right]
$$
where $\beta \in (0,1)$ is the household's subjective discount factor, $g(l_t)$ represents the disutility of supplying labor $l_t$, and $c_t$ is a CES aggregate of home and foreign goods.

Households' labor productivity $\{s_t\}_{t=0}^{\infty}$ is stochastic and is characterized by an $N$-state Markov chain that can take values $s_t \in S = \{s_1, \dots, s_N\}$ with transition probabilities given by $\gamma(s_{t+1} \mid s_t)$ and $\int s = 1$. Households rent their labor services, $l_t s_t$, to firms for a real wage $w_t$ per effective hour. $W_t$ denotes the corresponding nominal wage.

Agents rent their nominal asset portfolio $a_t$ to the asset market and earn a nominal return $i_t^a$. Households can hold multiple asset classes, including capital, real bonds, and nominal bonds denominated in domestic and foreign currencies. The nominal assets carried into the next period are given by
$$
a_{t+1} = b_{H,t+1} + e_t b_{F,t+1} + P_{H,t} \left( q_t k_{t+1} + \tilde{b}_{H,t+1} \right),
$$
where $b_{H,t+1}$ and $b_{F,t+1}$ denote nominal bonds denominated in domestic and foreign currency, respectively, $e_t$ is the nominal exchange rate, $q_t$ is the real price of capital in units of the home good, and $\tilde{b}_{H,t+1}$ are real bonds denominated in units of the home good. 

An unexpected change in the exchange rate $\Delta e_t$ therefore induces a one-time change in household wealth through the revaluation of foreign-currency debt, which affects household consumption.

Households take prices and wages as given. 
At time $t$, an agent faces the following nominal budget constraint:
$$
P_t c_t + a_{t+1} = (1 + i_t^a)a_t + W_t l_t s_t + d_t
$$
where $P_t$ is the price of a unit of consumption in terms of domestic currency, or the consumer price index (CPI), and $d_t$ are dividends distributed by firms.

In addition, households are subject to a borrowing constraint:
$$
a_{t+1} \geq \bar{a}
$$
which implies that assets carried into the next period cannot fall below the threshold $\bar{a}$.

##### **State variables and control variables**

In the household problem, the **state variables** at the beginning of a period are asset holdings $a$ and labor productivity $s$, while the **control variables** within the period are consumption $c$ and next period's asset holdings $a'$. 
In this model, labor supply is not a household choice variable in the Bellman problem. Instead, households supply labor services to firms at a wage that is set optimally by the union. As a result, $l_t$ is not treated as a control variable in the household's dynamic optimization problem.

##### **Shocks**

An idiosyncratic labor productivity shock $s$ is realized at the beginning of each period and becomes part of the household's state before decisions are made. The shock follows a Markov process, so households form expectations over the next-period productivity state $s'$ when solving the Bellman equation.


##### **Bellman equation**

The household's Bellman equation is given by:
$$
V(a,s;\Omega)
=
\max_{c \ge 0,\, a' \ge \bar{a}}
\left\{
u(c) - g(l)
+
\beta
\sum_{s' \in S}
\gamma(s' \mid s)
V(a', s'; \Omega')
\right\}
$$

subject to
$$
Pc + a' = (1 + i^a)a + Wls + \pi
$$

$$
\Omega' = \Upsilon(\Omega)
$$
where $\Omega(a,s)$ is the distribution on the space $X = \mathcal{A} \times \mathcal{S}$ of agents' asset holdings $a \in \mathcal{A}$ and labor productivity $s \in \mathcal{S}$ across the population, which, together with policy variables, determines equilibrium prices. $\Upsilon$ is an equilibrium object that specifies the evolution of the distribution $\Omega$.







#### **2. Why multiple stages?**

At the beginning of each period, the household observes the current state, including asset holdings and the realized productivity shock. Given this realization, total resources are determined and the household chooses consumption $c$ and next-period assets $a'$, subject to the budget constraint. Because of this constraint, the choice of $c$ determines $a'$.

**Therefore, although the Bellman equation in the paper appears to involve two control variables, $c$ and $a'$, I am inclined to interpret the problem as having one choice variable together with a non-trivial shock structure that generates multiple stages.**

Structuring the problem into stages clarifies the timing of shocks and decisions. In the DDSL framework, each stage becomes a module with well-defined perches, which makes the structure of the dynamic problem easier to understand, and facilitates its numerical solution and simulation.


#### **3. Stage decomposition**

Let 
$$
m_t = P_t c_t + a_{t+1} = (1 + i_t^a)a_t + W_t l_t s_t + \pi_t
$$
denote nominal market resources.

Once the productivity shock $s_t$ is realized, market resources $m_t$ are determined. 
Therefore, in the following analysis I use $m_t$ instead of $s_t$ as the relevant state variable, 
while leaving the other state variable unchanged.

Define 
$$
\psi_t = m_t - P_t c_t
$$
as assets after all actions are completed. (We do not use $k$ because $k$ denotes physical capital in this paper.) 

Next period's asset holdings are therefore given by
$$
a_{t+1} = \psi_t
$$

Both $\psi$ and $a$ are $\psi$-type variables, representing investable assets before returns are realized. In contrast, the $m$-type variables like $m$ represents spendable resources after returns.

##### **3.3 The Shock-free consumption Stage (cons-noshocks)**
With shocks in the preceding stage, the consumption stage has Arrival state $m$ and no shocks between Arrival and Decision, so $v_{\prec} = v_{\sim}$.

##### **3.4 The three-stage period**

The shocksonly–consnoshocks period is defined by the stage list  [shocks-only, $C(\check{m} \leftrightarrow m)$, cons-noshocks, disc]:

| Element | Transition | Action |
|--------|-----------|-------|
| shocks-only | $a \rightarrow \check{m}$ | shocks realize (no choice) |
| $\mathcal{C}(\check{m} \leftrightarrow m)$ | $\check{m} \rightarrow m$ | rename |
| cons-noshocks | $m \rightarrow \psi$ | choose $c$ |
| disc |  | apply $\beta$ |

##### **3.5 Connectors for Period**

The within-period connectors are specified above. For the between-period connector, the period ends after the disc stage (a passthrough), so the exit state is $\psi$. The shocks-only stage of the next period arrives with state $a$. Therefore, the between-period connector is
$\mathcal{C}(\psi \leftrightarrow a)$.

The pipeline is:
$$
a 
\xrightarrow{\text{shocks-only}} 
\check{m} 
\xrightarrow{\text{cons-noshocks}} 
\psi 
\xrightarrow{\text{disc}} 
\text{exit}
$$




#### **4. Comparison with the monolithic formulation**
##### **4.1 How does your staged decomposition relate to the Bellman equation you wrote in Section 1?**
The staged decomposition is mathematically equivalent to the Bellman equation presented in Section 1, but it separates the optimization into sequential stages. In the Bellman formulation,
$$
V(a_t,m_t;\Omega)
=
\max
\left\{
u(c_t) - g(l_t)
+
\beta
E\left[
V(a_{t+1}, m_{t+1}; \Omega_{t+1})
\right]
\right\}
$$
the current utility, continuation value, and expectation over future shocks appear in a single expression. In the staged decomposition, these components are implemented in separate modules: the shocks-only stage forms expectations over the productivity shock, the cons-noshocks stage determines $c_t$ and $a_{t+1}$, and the discounting stage applies the factor $\beta$ to the continuation value. Composed together, these stages reproduce the same value function as the monolithic Bellman equation.

##### **4.2 What does the decomposition make clearer that the monolithic formulation obscures?**
The staged decomposition clarifies that the realization of productivity determines market resources before the consumption decision is made, and that the budget constraint links consumption to next-period assets. In the monolithic Bellman equation, these timing relationships are less transparent.

##### **4.3 Are there alternative decompositions? Why did you choose this one?**
Yes. One can also consider a **consnoshocks–shocksonly** period, which is defined by the stage list  
[cons-noshocks, $C(\psi \leftrightarrow a)$, shocks-only, disc]:

| Element | Transition | Action |
|--------|-----------|-------|
| cons-noshocks | $m \rightarrow \psi$ | choose $c$ |
| $\mathcal{C}(\psi \leftrightarrow a)$ | $\psi \leftrightarrow a$ | rename |
| shocks-only | $a \rightarrow \check{m}$ | shocks realize (no choice) |
| disc |  | apply $\beta$ |

Under this decomposition, the between-period connector is $\mathcal{C}(\check{m} \leftrightarrow m)$.

I choose the **shocksonly–consnoshocks** period because it is more intuitive and consistent with the description of the model in the paper: households make consumption decisions after receiving wage income.
